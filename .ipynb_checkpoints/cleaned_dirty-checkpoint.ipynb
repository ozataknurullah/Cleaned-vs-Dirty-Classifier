# Data Preprocessing

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Standardize image size
IMG_SIZE = (224, 224)
```

```python
# Label mapping
label_map = {"cleaned": 0, "dirty": 1}
```

```python
# Function to preprocess images from a directory
def preprocess_images_from_directory(directory_path, label_map):
    data = []
    labels = []
    for label, value in label_map.items():
        label_path = os.path.join(directory_path, label)
        for image_file in os.listdir(label_path):
            image_path = os.path.join(label_path, image_file)
            img = cv2.imread(image_path)
            if img is not None:
                img = cv2.resize(img, IMG_SIZE)  # Resize image
                img = img / 255.0  # Normalize image
                data.append(img)
                labels.append(value)
    return np.array(data, dtype="float32"), np.array(labels, dtype="int")
```

```python
# Function to preprocess test images using CSV
def preprocess_test_images_from_csv(test_images_path, labels_csv_path, label_map):
    data = []
    labels = []
    labels_df = pd.read_csv(labels_csv_path)
    for _, row in labels_df.iterrows():
        image_name = str(row["id"]).zfill(4) + ".jpg"  # Ensure leading zeros are kept
        label = row["label"]
        image_path = os.path.join(test_images_path, image_name)
        img = cv2.imread(image_path)
        if img is None:
            print(f"Failed to load image: {image_name}")
            continue
        img = cv2.resize(img, IMG_SIZE)  # Resize image
        img = img / 255.0  # Normalize image
        data.append(img)
        labels.append(label_map[label])
    return np.array(data, dtype="float32"), np.array(labels, dtype="int")
```

```python
# Configure data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
# Preprocess train data
X_train, y_train = preprocess_images_from_directory(train_dataset_path, label_map)
print("Train data shape:", X_train.shape)
print("Train labels shape:", y_train.shape)
```

```python
# Preprocess test data
X_test, y_test = preprocess_test_images_from_csv(test_dataset_path, labels_csv_path, label_map)
print("Test data shape:", X_test.shape)
print("Test labels shape:", y_test.shape)
```

# Model Development

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, GlobalAveragePooling2D
```

```python
# Build the CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')  # 2 classes: cleaned and dirty
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

# Evaluating Model Performance

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
# Predict on test data
y_pred = np.argmax(model.predict(X_test), axis=1)
```

```python
# Classification report
print(classification_report(y_test, y_pred, target_names=["cleaned", "dirty"]))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["cleaned", "dirty"], yticklabels=["cleaned", "dirty"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
